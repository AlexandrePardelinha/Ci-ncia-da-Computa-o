```python
import networkx as nx
from karateclub import DeepWalk
import matplotlib.pyplot as plt

g = nx.newman_watts_strogatz_graph(10, 4, 0.2)

model = DeepWalk(dimensions = 2)
model.fit(g)
embedding = model.get_embedding()
```

```python
nx.draw_networkx(g, nx.layout.circular_layout(g))
```

```python
plt.plot(*embedding.T, '.')

for i, xy in enumerate(embedding):
    plt.annotate(i, xy)
```
