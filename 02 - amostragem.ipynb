```python
import networkx as nx
import numpy as np
```

```python
G = nx.gnm_random_graph(10, 10)
nx.draw_networkx(G, nx.layout.circular_layout(G))
```

```python
nodes = np.random.choice(G.nodes(), 5, replace=False)
H = nx.induced_subgraph(G, nodes)
print(nodes)
nx.draw_networkx(H, nx.layout.circular_layout(H))
```

```python
indexes = np.random.choice(len(G.edges()), 5, replace=False)
edges = np.array(G.edges())[indexes]
H = nx.from_edgelist(edges)
nx.draw_networkx(H, nx.layout.circular_layout(H))
```
