```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import math
```

```python
n = 10
p = 0.5
```

```python
np.log(n)/n
```

```python
arestas = []
for i in range(n):
    for j in range(i+1, n):
        if np.random.random() < p:
            arestas.append((i,j))

print(len(arestas))            
G = nx.from_edgelist(arestas)
nx.draw_networkx(G, nx.layout.circular_layout(G))
```

```python
empirico = []
teorico = []
for p in np.linspace(0,1,11):
    m = []
    for i in range(100):
        m.append(len(nx.erdos_renyi_graph(n, p).edges()))
    empirico.append(np.mean(m))
    teorico.append(math.comb(n,2) * p)
```

```python
plt.plot(empirico)
plt.plot(teorico)
```

```python
n = 10
empirico = []
for p in np.linspace(0,1,21):
    is_connected = []
    for i in range(100):
        G = nx.erdos_renyi_graph(n, p)
        is_connected.append(nx.is_connected(G))
    empirico.append(np.mean(is_connected))
```

```python
plt.plot(np.linspace(0,1,21), empirico, '-s')
plt.vlines(math.log(n)/n, 0, 1)
```

```python
y1 = []
y2 = []
for p in np.linspace(0,1,21):
    medida1 = []
    medida2 = []
    for i in range(10):
        G = nx.watts_strogatz_graph(100, 10, p)
        medida1.append(nx.average_shortest_path_length(G))
        medida2.append(nx.average_clustering(G))
        
    y1.append(np.mean(medida1))
    y2.append(np.mean(medida2))

plt.subplot(1,2,1)
plt.plot(np.linspace(0,1,21), y1)

plt.subplot(1,2,2)
plt.plot(np.linspace(0,1,21), y2)


#, nx.average_shortest_path_length(G)
```
